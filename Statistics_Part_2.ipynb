

 **1. What is hypothesis testing in statistics?**

**Hypothesis testing** is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence to support a specific hypothesis or claim.


**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

* **Null Hypothesis (H₀):** A statement that there is no effect or difference (e.g., μ = 0).
* **Alternative Hypothesis (H₁ or Ha):** A statement that contradicts the null, indicating an effect or difference (e.g., μ ≠ 0).
* **Difference:** The null assumes no change; the alternative suggests a meaningful difference or effect.


**3. What is the significance level in hypothesis testing, and why is it important?**

The **significance level (α)** is the probability of rejecting the null hypothesis when it is true (Type I error). Common values are 0.05 or 0.01. It defines the threshold for determining whether results are statistically significant.


**4. What does a P-value represent in hypothesis testing?**

A **P-value** is the probability of obtaining a result at least as extreme as the observed one, assuming the null hypothesis is true.


**5. How do you interpret the P-value in hypothesis testing?**

* **P-value ≤ α:** Reject the null hypothesis (statistically significant).
* **P-value > α:** Fail to reject the null hypothesis (not significant).


**6. What are Type 1 and Type 2 errors in hypothesis testing?**

* **Type I Error (False Positive):** Rejecting a true null hypothesis.
* **Type II Error (False Negative):** Failing to reject a false null hypothesis.


**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

* **One-tailed test:** Tests for deviation in one direction (greater or lesser).
* **Two-tailed test:** Tests for deviation in both directions (not equal).


**8. What is the Z-test, and when is it used in hypothesis testing?**

A **Z-test** is used when the population standard deviation is known and the sample size is large (n > 30), typically for comparing means or proportions.


**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

**Z-score formula:**

$$
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
$$

It measures how many standard deviations the sample mean is from the population mean under H₀.


**10. What is the T-distribution, and when should it be used instead of the normal distribution?**

The **T-distribution** is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It has heavier tails to account for increased variability.


**11. What is the difference between a Z-test and a T-test?**

* **Z-test:** Known population standard deviation, large sample.
* **T-test:** Unknown population standard deviation, small sample.


**12. What is the T-test, and how is it used in hypothesis testing?**

A **T-test** compares sample means to test hypotheses about population means. Types include:

* **One-sample T-test:** Compare sample mean to population mean.
* **Independent T-test:** Compare means of two independent groups.
* **Paired T-test:** Compare means from the same group at different times.


**13. What is the relationship between Z-test and T-test in hypothesis testing?**

Both are used to test hypotheses about means. The **Z-test** is a special case of the **T-test** when the sample size is large and the population standard deviation is known.


**14. What is a confidence interval, and how is it used to interpret statistical results?**

A **confidence interval (CI)** is a range of values within which the population parameter is expected to lie with a certain level of confidence (e.g., 95%). It reflects the precision of the estimate.


**15. What is the margin of error, and how does it affect the confidence interval?**

The **margin of error** is the amount added and subtracted from the sample estimate to create the confidence interval. A larger margin results in a wider, less precise interval.


**16. How is Bayes' Theorem used in statistics, and what is its significance?**

**Bayes' Theorem** updates the probability of a hypothesis based on new evidence.
Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

It’s significant in decision-making under uncertainty and in fields like machine learning and diagnostics.


**17. What is the Chi-square distribution, and when is it used?**

The **Chi-square distribution** is used in tests of independence and goodness-of-fit for categorical data. It’s right-skewed and based on squared deviations.


**18. What is the Chi-square goodness of fit test, and how is it applied?**

This test checks whether observed frequencies match expected frequencies for a categorical variable. It compares observed vs. expected counts using:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$


**19. What is the F-distribution, and when is it used in hypothesis testing?**

The **F-distribution** is used to compare two variances. It is commonly used in **ANOVA** and regression analysis.


**20. What is an ANOVA test, and what are its assumptions?**

**ANOVA (Analysis of Variance)** tests whether there are significant differences between the means of three or more groups.
**Assumptions:**

1. Independence of observations
2. Normal distribution of the groups
3. Equal variances (homoscedasticity)


**21. What are the different types of ANOVA tests?**

1. **One-way ANOVA:** One independent variable
2. **Two-way ANOVA:** Two independent variables
3. **Repeated Measures ANOVA:** Same subjects under different conditions


**22. What is the F-test, and how does it relate to hypothesis testing?**

An **F-test** evaluates whether two population variances are equal or if group means differ significantly (as in ANOVA). It uses the F-distribution and is central to variance-based testing.

3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1)
        The outcome of the Bernoulli trial
    p : float
        The probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of the outcome
    """
    if x not in [0, 1]:
        return 0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
```

9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (array-like): Input numerical dataset

    Returns:
    z_scores (numpy array): Calculated Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be calculated.")

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8,4))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

# Example usage:
data = [10, 12, 23, 23, 16, 23, 21, 16]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

14. E Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score of data point x in the dataset.

    Parameters:
    data (array-like): Dataset
    x (float): Data point to calculate Z-score for

    Returns:
    z (float): Z-score of x
    """
    mean = np.mean(data)
    std = np.std(data)
    if std == 0:
        raise ValueError("Standard deviation is zero, cannot calculate Z-score.")
    z = (x - mean) / std
    return z

# Example dataset and data point
data = [12, 15, 14, 10, 8, 13, 17, 14, 12, 11]
x = 16  # data point to calculate Z-score for

# Calculate Z-score
z_score = calculate_z_score(data, x)
print(f"Z-score of data point {x}: {z_score:.2f}")

# Plot standard normal distribution
x_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(x_vals, 0, 1)

plt.plot(x_vals, pdf_vals, label='Standard Normal PDF')
plt.fill_between(x_vals, 0, pdf_vals, alpha=0.1)

# Mark the Z-score on the plot
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title('Standard Normal Distribution with Z-score')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test on sample data.

    Parameters:
    - sample_data: array-like sample data
    - pop_mean: hypothesized population mean (H0)
    - pop_std: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - reject_null: True if reject H0, False otherwise
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decide whether to reject null hypothesis
    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

# Example sample data
sample = [105, 102, 98, 100, 110, 95, 99, 101, 103, 97]

# Known population parameters
population_mean = 100
population_std = 10
alpha = 0.05

z_stat, p_value, reject = z_test(sample, population_mean, population_std, alpha)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print("Reject null hypothesis:" if reject else "Fail to reject null hypothesis.")
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.

    Parameters:
    - data: array-like dataset
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean
    df = n - 1  # degrees of freedom

    # t-score for two-tailed test
    t_score = stats.t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_score * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset
data = [12, 15, 14, 10, 8, 13, 17, 14, 12, 11]

# Calculate 95% confidence interval
ci_lower, ci_upper = confidence_interval(data, confidence=0.95)

print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies within this interval.")
```

17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

# Step 1: Generate normal data
np.random.seed(42)  # for reproducibility
mean = 50
std_dev = 10
n = 100

data = np.random.normal(loc=mean, scale=std_dev, size=n)

# Step 2: Calculate confidence interval for the mean
confidence = 0.95
sample_mean = np.mean(data)
sem = stats.sem(data)  # Standard error of the mean
df = n - 1  # degrees of freedom
t_score = stats.t.ppf((1 + confidence) / 2, df)

# Margin of error
margin_of_error = t_score * sem

# Confidence interval bounds
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 3: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("\nInterpretation: We are 95% confident that the true population mean lies within this interval.")
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

# Parameters
population_mean = 100  # μ (known population mean)
population_std = 15     # σ (known population standard deviation)

# Generate a sample dataset (normally distributed)
np.random.seed(0)
sample = np.random.normal(loc=103, scale=15, size=50)  # mean=103, std=15, n=50
sample_mean = np.mean(sample)
sample_size = len(sample)

# Step 1: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 2: Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 3: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret results (α = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Hypothesis test conclusion
    if p_value < alpha:
        print("Reject the null hypothesis: sample mean differs significantly from population mean.")
    else:
        print("Fail to reject the null hypothesis: no significant difference from population mean.")

# Example usage:
np.random.seed(1)
sample_data = np.random.normal(loc=105, scale=10, size=30)  # Sample with mean around 105
population_mean = 100
population_std = 12

one_sample_z_test(sample_data, population_mean, population_std)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Critical z values for two-tailed test
    z_crit = norm.ppf(1 - alpha / 2)

    # Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3)

    # Plot Z-statistic
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')

    plt.title('Two-Tailed Z-Test: Standard Normal Distribution')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Interpretation
    if abs(z_stat) > z_crit:
        print("Decision: Reject the null hypothesis (significant difference).")
    else:
        print("Decision: Fail to reject the null hypothesis (no significant difference).")

# Example usage:
np.random.seed(123)
sample = np.random.normal(loc=102, scale=15, size=40)  # Sample data
population_mean = 100
population_std = 15
alpha = 0.05

two_tailed_z_test(sample, population_mean, population_std, alpha)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    """
    Visualizes Type 1 and Type 2 errors for a one-sample Z-test.

    Parameters:
    - mu0: Mean under null hypothesis (H0)
    - mu1: Mean under alternative hypothesis (H1)
    - sigma: Population standard deviation
    - alpha: Significance level (Type 1 error rate)
    - n: Sample size
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (two-tailed)
    z_crit = norm.ppf(1 - alpha / 2)

    # Critical value in original units
    crit_low = mu0 - z_crit * se
    crit_high = mu0 + z_crit * se

    # Define range for x axis
    x_min = mu0 - 4 * se
    x_max = mu1 + 4 * se
    x = np.linspace(x_min, x_max, 1000)

    # Distributions under H0 and H1
    pdf_h0 = norm.pdf(x, mu0, se)
    pdf_h1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(12, 7))

    # Plot H0 distribution
    plt.plot(x, pdf_h0, label='Null Hypothesis (H0)', color='blue')
    # Plot H1 distribution
    plt.plot(x, pdf_h1, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error region (reject H0 when true)
    plt.fill_between(x, 0, pdf_h0, where=(x < crit_low), color='red', alpha=0.3, label='Type 1 Error (α)')
    plt.fill_between(x, 0, pdf_h0, where=(x > crit_high), color='red', alpha=0.3)

    # Shade Type 2 error region (fail to reject H0 when H1 true)
    plt.fill_between(x, 0, pdf_h1, where=( (x >= crit_low) & (x <= crit_high) ), color='orange', alpha=0.3, label='Type 2 Error (β)')

    # Mark critical values
    plt.axvline(crit_low, color='red', linestyle='--', label=f'Critical Value = {crit_low:.2f}')
    plt.axvline(crit_high, color='red', linestyle='--')

    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Calculate beta (Type 2 error)
    beta = norm.cdf(crit_high, loc=mu1, scale=se) - norm.cdf(crit_low, loc=mu1, scale=se)
    print(f"Significance level (Type 1 error, α): {alpha}")
    print(f"Type 2 error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=0, mu1=1.5, sigma=1, alpha=0.05, n=30)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample: array-like, the data sample
    - confidence: float, confidence level (default 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
np.random.seed(0)
data = np.random.normal(loc=50, scale=8, size=40)

ci_lower, ci_upper = confidence_interval(data, confidence=0.95)
print(f"Sample mean: {np.mean(data):.2f}")
print(f"95% Confidence interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, sensitivity, specificity, test_result='positive'):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(Disease), prior probability of having the disease
    - sensitivity: P(Test+ | Disease), true positive rate
    - specificity: P(Test- | No Disease), true negative rate
    - test_result: 'positive' or 'negative' test result

    Returns:
    - posterior: P(Disease | Test result)
    """
    # Complementary probabilities
    prior_neg = 1 - prior  # P(No Disease)
    false_positive_rate = 1 - specificity  # P(Test+ | No Disease)
    false_negative_rate = 1 - sensitivity  # P(Test- | Disease)

    if test_result == 'positive':
        # Bayes theorem for positive test
        numerator = sensitivity * prior
        denominator = numerator + false_positive_rate * prior_neg
    elif test_result == 'negative':
        # Bayes theorem for negative test
        numerator = false_negative_rate * prior
        denominator = numerator + specificity * prior_neg
    else:
        raise ValueError("test_result must be 'positive' or 'negative'")

    posterior = numerator / denominator
    return posterior

# Parameters
prior = 0.01          # 1% disease prevalence
sensitivity = 0.99    # 99% true positive rate
specificity = 0.95    # 95% true negative rate

# Calculate posterior probability given a positive test
posterior_positive = bayes_theorem(prior, sensitivity, specificity, test_result='positive')
print(f"Probability of having the disease given a positive test: {posterior_positive:.4f}")

# Calculate posterior probability given a negative test
posterior_negative = bayes_theorem(prior, sensitivity, specificity, test_result='negative')
print(f"Probability of having the disease given a negative test: {posterior_negative:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test given observed data.

    Parameters:
    - observed: 2D numpy array of observed counts

    Returns:
    - expected: 2D numpy array of expected counts
    """
    # Row sums and column sums
    row_sums = observed.sum(axis=1).reshape(-1, 1)  # column vector
    col_sums = observed.sum(axis=0).reshape(1, -1)  # row vector
    total = observed.sum()

    # Calculate expected frequencies = (row_sum * col_sum) / total
    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data
observed = np.array([[30, 10],
                     [20, 40]])

expected = calculate_expected_frequencies(observed)

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```
